```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 105, 1006)
```

```python
x = np.round(x, 1) # 소숫점 문제가 존재해서 첫째자리를 기준으로 반올림
```

```python
y = np.cos(x)
```

```python
plt.plot(x, y)
```

```python
# Test data와 train 데이터로 나눠준다.
```

```python
df = pd.DataFrame(data=y, index=x, columns=["cosine_wave"])
```

```python
df
```

```python
df.index
```

```python
train = df.iloc[:957]
test = df.iloc[957:1007]
```

```python
train
```

```python
test
```

```python
# scale data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
s_train = scaler.transform(train)
```

```python
s_train.shape
```

```python
s_test = scaler.transform(test)
```

```python
s_test.shape
```

```python
# Time series Generator
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 30
generator = TimeseriesGenerator(s_train, s_train, length=length, batch_size=1)
```

```python
generator[0]
```

```python
generator[1]
```

```python
# model 생성
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.SimpleRNN(100, input_shape=(49,1)))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(generator, epochs=4)
#generator = TimeseriesGenerator(s_train, s_train, length=length, batch_size=1)
```

```python
s_g_test = TimeseriesGenerator(s_test, s_test, length=length, batch_size=1)
```

여기서 length를 정확하게 맞추지 못해서 error가 발생했다.
50, 49, 48로 해보니까 48개가 맞았다

```python
predict_test = model.predict(s_g_test)
```

```python
true_predict = scaler.inverse_transform(predict_test)
```

```python
true_predict.shape
```

```python
test.shape
```

```python
test2 = test.drop(test.index[0:length])
```

```python
test2["predict_test_simplernn"] = true_predict
```

```python
test2
```

```python
plt.plot(test2.index, test2.cosine_wave, linewidth = 5.0)
plt.plot(test2.index, test2.predict_test_simplernn, linewidth = 2.0)
```

```python
# 포인트는 lenght 를 몇으로 하냐, 에 달린 듯
# 처음엔 수업내용 sin 처럼 50으로 두었으나
# 이번 모델은 shape이 (49, 1)이어서 그보다 작은 30으로 두고 돌려보았음
# 그 결과 예상치가 잘 나옴
```
