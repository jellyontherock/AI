```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('../input/tiktok-google-play-store-review/tiktok_google_play_reviews.csv')
df.head()
```

```python
df.info()
```

```python
df2 = df[df['reviewCreatedVersion'].isnull() == True].index
df = df.drop(df2)
df.info()
```

```python
df = pd.DataFrame(df)
```

```python
df = df.astype({'reviewCreatedVersion':'str'})
df
```

```python
df['v1'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[0])
df['v2'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[1])
df['v3'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[2])
df.head()
```

```python
df = df.astype({'v1':'int', 'v2':'int', 'v3':'int'})
```

```python
df = df.sort_values(by = ['score', 'v1', 'v2', 'v3'], ascending = [True, False, False, False])
df.head()
```

```python
fig = px.histogram(df, x="score")
fig.show()
```

```python
grouped = df['v1'].groupby(df['score'])
grouped.count()
```

```python
grouped = df[df['score']==1].groupby(df['v1'])
s1 = grouped.count()
v1 = s1['v1']
v1 = pd.DataFrame(v1)

grouped = df[df['score']==2].groupby(df['v1'])
s2 = grouped.count()
v2 = s2['v1']
v2 = pd.DataFrame(v2)

grouped = df[df['score']==3].groupby(df['v1'])
s3 = grouped.count()
v3 = s3['v1']
v3 = pd.DataFrame(v3)

grouped = df[df['score']==4].groupby(df['v1'])
s4 = grouped.count()
v4 = s4['v1']
v4 = pd.DataFrame(v4)

grouped = df[df['score']==5].groupby(df['v1'])
s5 = grouped.count()
v5 = s5['v1']
v5 = pd.DataFrame(v5)
```

```python
table = range(0,24)
just_table = pd.DataFrame(table)
just_table['v_num'] = just_table[0]
score_table = just_table.drop(just_table.columns[0], axis = 1)
score_table.head()
```

```python
score_table = pd.concat([score_table, v1, v2, v3, v4, v5], axis=1)
```

```python
score_table.columns = ['v_num', 'score1', 'score2', 'score3', 'score4', 'score5']
```
