```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('../input/tiktok-google-play-store-review/tiktok_google_play_reviews.csv')
df.head()
```

```python
df.info()
```

```python
df2 = df[df['reviewCreatedVersion'].isnull() == True].index
df = df.drop(df2)
df.info()
```

```python
df = pd.DataFrame(df)
```

```python
df = df.astype({'reviewCreatedVersion':'str'})
df
```

```python
df['v1'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[0])
df['v2'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[1])
df['v3'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[2])
df.head()
```

```python
df = df.astype({'v1':'int', 'v2':'int', 'v3':'int'})
```

```python
df = df.sort_values(by = ['score', 'v1', 'v2', 'v3'], ascending = [True, False, False, False])
df.head()
```

```python
fig = px.histogram(df, x="score")
fig.show()
```

```python
grouped = df['v1'].groupby(df['score'])
grouped.count()
```

```python
grouped = df['score'].groupby(df['v1'])
grouped.count()
```
