```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('../input/tiktok-google-play-store-review/tiktok_google_play_reviews.csv')
df.head()
```

```python
df.info()
```

```python
df2 = df[df['reviewCreatedVersion'].isnull() == True].index
df = df.drop(df2)
df.info()
```

```python
df = pd.DataFrame(df)
```

```python
df = df.astype({'reviewCreatedVersion':'str'})
df
```

```python
df['v1'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[0])
df['v2'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[1])
df['v3'] = df['reviewCreatedVersion'].apply(lambda x: x.split('.')[2])
df.head()
```

```python
df = df.astype({'v1':'int', 'v2':'int', 'v3':'int'})
```

```python
df = df.sort_values(by = ['score', 'v1', 'v2', 'v3'], ascending = [True, False, False, False])
df.head()
```

```python
fig = px.histogram(df, x="score")
fig.show()
```

```python
score_1 = df[df['score'] == 1]
score_2 = df[df['score'] == 2]
score_3 = df[df['score'] == 3]
score_4 = df[df['score'] == 4]
score_5 = df[df['score'] == 5]
```

```python
version = ['19', '20', '21', '22', '23']
score1 = score_1['v1'].value_counts()
score2 = score_2['v1'].value_counts()
score3 = score_3['v1'].value_counts()
score4 = score_4['v1'].value_counts()
score5 = score_5['v1'].value_counts()

sv1 = go.Bar(
   x = version,
   y = score1,
    name = '1'
)

sv2 = go.Bar(
   x = version,
   y = score2,
    name = '2'
)

sv3 = go.Bar(
   x = version,
   y = score3,
    name = '3'
)

sv4 = go.Bar(
   x = version,
   y = score4,
    name = '4'
)

sv5 = go.Bar(
   x = version,
   y = score5,
    name = '5'
)


data = [sv1, sv2, sv3, sv4, sv5]
layout = go.Layout(barmode = 'group')
fig2 = go.Figure(data = data, layout = layout)
fig2.update_layout(autosize=True, title='Score and Version')
fig2.show()
```
