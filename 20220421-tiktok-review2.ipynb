```python
df = pd.read_csv("../input/tiktok-google-play-store-review/tiktok_google_play_reviews.csv")
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
df = df.drop([ 'replyContent', 'repliedAt'], axis = 1)
df.columns
```

```python
df['date'] = df['at'].apply(lambda x: x.split(' ')[0])
df.head()
```

```python
df['count'] = [1] * df.shape[0]
```

```python
good_df = df[df['score'] > 2]
bad_df = df[df['score'] < 3]
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
bad_df = bad_df[bad_df['content'].notna()]
```

```python
bad_df.info()
```

```python
bad_df = bad_df.sort_values(by = 'thumbsUpCount', ascending = False)
bad_df = bad_df.sort_values(by = 'score', ascending = True)
bad_df.head()
```
